**concatinating DataFrames**

```python
import pandas as pd
a=['fwf','fws','gdf','fde']
b=['m','f','m','f']
c=[10000,20000,30000,40000]
d=pd.DataFrame({'Name':a,'Gender':b,'Salary':c})
d
```

```python
import pandas as pd
a2=['fwf','fws','gdf','fde','uyh']
b2=['f','m','f','m','m']
c2=[300,400,200,500,979]
d2=pd.DataFrame({'Name':a2,'Gender':b2,'fees':c2})
d2
```

```python
df=pd.concat([d,d2],axis=1)
df
```

**merge**

```python
df2=pd.merge(d,d2,how='outer',on='Name')
df2
```

```python
df3=pd.merge(d,d2,how='inner',on='Name')
df3
```

```python
df4=pd.merge(d,d2,how='right',on='Name')
df4
```
